```python
import sqlite3
import pandas as pd

# Example: files_list contains your Excel files
# files_list = ["Cleaned_Data_Final_pe.xlsx", ...]

# Step 1: Read each file and store in a list
li = []
for f in files_list:
    df = pd.read_excel(f)
    li.append(df)

# Step 2: Create and connect to the database
conn = sqlite3.connect('hajj_companies.db')
cursor = conn.cursor()

# Step 3: Create the table 'agencies' with updated columns
cursor.execute('''
CREATE TABLE IF NOT EXISTS agencies (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    hajj_company_ar TEXT,
    hajj_company_en TEXT,
    formatted_address TEXT,
    city TEXT,
    country TEXT,
    email TEXT,
    contact_info TEXT,
    rating_reviews TEXT,
    is_authorized BOOLEAN
)
''')

conn.commit()

# Step 4 (optional): Insert data from the DataFrames into the table
for df in li:
    for _, row in df.iterrows():
        cursor.execute('''
            INSERT INTO agencies (
                hajj_company_ar, hajj_company_en, formatted_address, city, country,
                email, contact_info, rating_reviews, is_authorized
            ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)
        ''', (
            row.get("Hajj Company (Arabic)"),
            row.get("Hajj Company (English)"),
            row.get("formattedAddress"),
            row.get("City"),
            row.get("Country"),
            row.get("email"),
            row.get("Contact_Info"),
            row.get("Rating Reviews"),
            row.get("is_autorized")
        ))

conn.commit()
conn.close()
```

```python
import sqlite3
import pandas as pd

# Connect to the database
conn = sqlite3.connect('hajj_companies.db')

# Rename columns and insert data
for df in li:
    df.rename(columns={
        'Hajj Company (Arabic)': 'hajj_company_ar',
        'Hajj Company (English)': 'hajj_company_en',
        'formattedAddress': 'formatted_address',
        'City': 'city',
        'Country': 'country',
        'email': 'email',
        'Contact_Info': 'contact_info',
        'Rating Reviews': 'rating_reviews',
        'is_autorized': 'is_authorized'
    }, inplace=True)

    # Insert into the database
    df.to_sql('agencies', conn, if_exists='append', index=False)

# Commit and close **after all inserts**
conn.commit()
conn.close()

print("Data inserted into the database successfully.")
```
