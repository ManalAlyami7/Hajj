```python
from google.colab import files
import pandas as pd
from google.colab import data_table

#Step 1: Upload files from your local computer
uploaded = files.upload()
files_list = list(uploaded.keys())
```

```python
import sqlite3

# Step 2: Read each file and store in a list
li = []
for f in files_list:
    df = pd.read_excel(f)
    li.append(df)

# Create and connect to the database (if it doesn't exist, it will be created)
conn = sqlite3.connect('hajj_companies.db')
cursor = conn.cursor()

# Create the table 'agencies' if it does not exist
cursor.execute('''
CREATE TABLE IF NOT EXISTS agencies (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    hajj_company_ar TEXT,
    hajj_company_en TEXT,
    city TEXT,
    country TEXT,
    email TEXT,
    is_authorized BOOLEAN
)
''')

# Commit the changes to the database
conn.commit()
```

```python
# Rename columns in the DataFrame to match the database table
for df in li:
    df.rename(columns={
        'Hajj Company (Arabic)': 'hajj_company_ar',
        'Hajj Company (English)': 'hajj_company_en',
        'City': 'city',
        'Country': 'country',
        'email': 'email',
        'is_autorized': 'is_authorized' # Correcting the typo
    }, inplace=True)
    df.to_sql('agencies', conn, if_exists='append', index=False)

# Commit the changes and close the connection
conn.commit()
conn.close()

print("Data inserted into the database successfully.")
```

```python
# Connect to the database
conn = sqlite3.connect('hajj_companies.db')
cursor = conn.cursor()

# Select all data from the agencies table
cursor.execute("SELECT * FROM agencies LIMIT 5") # Limiting to 5 for display purposes
rows = cursor.fetchall()

# Print the results
for row in rows:
    print(row)

# Close the connection
conn.close()
```
